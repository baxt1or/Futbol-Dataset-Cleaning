```python
import pandas as pd 
```

```python
df = pd.read_excel("./datasets/data_clean.xlsx")

df
```

```python
df_copy = df.loc[9:].reset_index()

columns = df_copy.loc[0].tolist()[2:9]

df_copy = df_copy[df_copy.columns[2:9]]

old_columns = df_copy.columns

new_columns = {old_cols:cols for old_cols, cols in zip(old_columns, columns)}

df_copy = df_copy.rename(columns=new_columns)

df_copy = df_copy.loc[1:].reset_index()

df_copy
```

```python
positions = df_copy.loc[:, 'positions'].tolist()

positions
```

```python
import ast
from typing import defaultdict

positions = df_copy.loc[:, 'positions'].tolist()

data_list = []

for entry in positions:
    content = ast.literal_eval(entry) 

    if not content:  
        continue


    if isinstance(content, list):  
        data_list.extend(content)
    elif isinstance(content, dict): 
        data_list.append(content)


data_collection = defaultdict(list)

for entry in data_list:
    if isinstance(entry, dict):
       for key, val in dict(entry).items():
          data_collection[key].append(val)


result_df = pd.DataFrame(data_collection)

result_df
```

```python
import ast

df_copy["position_id"] = df_copy["positions"].apply(lambda data: [entry["position_id"] for entry in ast.literal_eval(data)])

columns = df_copy.columns.tolist()

columns.remove('cards')
columns.remove('positions')

df_copy = df_copy[columns]

df_copy
```

```python
df_copy["position_id"] = df_copy["position_id"].apply(lambda x: x if isinstance(x, list) else [])

df_exploded = df_copy.explode("position_id").reset_index(drop=True)

df_exploded
```

```python
final_df = pd.merge(df_exploded,result_df, on="position_id", how="inner")

final_df
```
